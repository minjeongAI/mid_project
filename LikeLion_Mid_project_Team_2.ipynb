```python
# 열 이름 수정 함수
def modify_column_name(name):
    # 점(.)이 있는 경우
    if '.' in name:
        # 첫 번째 점(.)에서 나누고 첫 번째 부분(연도)을 가져오기
        year = name.split('.')[0]  # '2017', '2018', ...
        # 연도와 나머지 부분을 조합하여 새로운 이름 생성
        return f"{year}_{name.split('_', 1)[1]}"
    else:
        # 점이 없거나 문자열만 있는 경우 원래 이름 반환
        return name

# 모든 열 이름에 함수 적용
modified_columns = [modify_column_name(col) for col in merged_df.columns]

merged_df.columns = modified_columns
```
